Now we define a few helper functions - it will make the process of building models much easier to understand

```python
def filter_data(input_df, train_x_df): # Applies the same changes for multiple df
    threshold = 0.00 # Selecting the variance threshold - very low (zero)
    sel = VarianceThreshold(threshold=threshold)
    transformed_features = sel.fit_transform(train_x_df)
    selected_columns = sel.get_support(indices=True)
    selected_columns = input_df.columns[selected_columns]
    selected_df = input_df[selected_columns]
    selected_df # df with the low variance columns removed
    return selected_df
```

```python
def calculate_and_save_min_max(input_df, train_x_df=None, save_path='min_max_values.csv'):
    if train_x_df is None:
        if os.path.exists(save_path):
            min_max_df = pd.read_csv(save_path, index_col=0)
            return min_max_df
        else:
            raise ValueError("No training set provided and no saved min-max values found.")
    
    min_max_values = {}
    for col in input_df.columns:
        min_value = train_x_df[col].min()
        max_value = train_x_df[col].max()
        min_max_values[col] = {'min': min_value, 'max': max_value}
    
    min_max_df = pd.DataFrame(min_max_values).T
    min_max_df.to_csv(save_path)
    return min_max_df
```

```python
def replace_outliers_with_extremes(input_df, train_x_df=None, min_max_file='min_max_values.csv'):
    if train_x_df is None:
        min_max_df = calculate_and_save_min_max(input_df, train_x_df, save_path=min_max_file)
    else:
        min_max_df = calculate_and_save_min_max(input_df, train_x_df)

    df_copy = input_df.copy()
    for col in df_copy.columns:
        lower_bound = min_max_df.at[col, 'min']
        upper_bound = min_max_df.at[col, 'max']
        
        lowest_non_outlier = df_copy[(df_copy[col] >= lower_bound) & (df_copy[col] <= upper_bound)][col].min()
        df_copy.loc[df_copy[col] < lower_bound, col] = lowest_non_outlier
        
        highest_non_outlier = df_copy[(df_copy[col] >= lower_bound) & (df_copy[col] <= upper_bound)][col].max()
        df_copy.loc[df_copy[col] > upper_bound, col] = highest_non_outlier
    
    return df_copy
```

```python
def normalize_data(input_df):
    scaler = MinMaxScaler()
    normalized_data = scaler.fit_transform(input_df)
    normalized_df = pd.DataFrame(normalized_data, columns=input_df.columns)
    return normalized_df
```

```python
# Create a function to plot ROC curve
def plot_roc(model, X_test, y_test, title = 'Receiver Operating Characteristic'):
    y_score = model.predict_proba(X_test)[:,1]
    fpr, tpr, _ = roc_curve(y_test, y_score)
    roc_auc = auc(fpr, tpr)
    plt.figure()
    plt.plot(fpr, tpr, color='darkorange', lw=2, label='ROC curve (area = %0.2f)' % roc_auc)
    plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
    plt.xlim([0.0, 1.0])
    plt.ylim([0.0, 1.05])
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title(title)
    plt.legend(loc="lower right")
    plt.show()
```

```python
def calculate_metrics(actual, predictions, title = 'Receiver Operating Characteristic (ROC) Curve'):
    fpr, tpr, thresholds = roc_curve(actual, predictions)
    roc_auc = auc(fpr, tpr)

    plt.figure(figsize=(8, 6))
    plt.plot(fpr, tpr, color='blue', lw=2, label='ROC curve (area = %0.2f)' % roc_auc)
    plt.plot([0, 1], [0, 1], color='red', lw=2, linestyle='--')
    plt.xlim([0.0, 1.0])
    plt.ylim([0.0, 1.05])
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title(title)
    plt.legend(loc='lower right')
    plt.show()
```
